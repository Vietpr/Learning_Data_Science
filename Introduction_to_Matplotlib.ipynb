### Pyplot API vs Object-Oriented API
* Pyplot API -> Quickly
* OO API -> Advanced

Pyplot API

```python
x = np.linspace(0, 10, 1000)  #  to ra mt mng x cha 1000 im nm trong khong t 0 n 10. np.linspace() to ra cc gi tr cch u nhau trong khong  cho.
plt.plot(x, np.sin(x), color = "blue");
plt.xlabel("Bien(X)") # Label: dn nhn cho trc
plt.ylabel("Sin(X)")
plt.xlim([0,4])
```

```python
x = np.linspace(0, 10, 1000)  #  to ra mt mng x cha 1000 im nm trong khong t 0 n 10. np.linspace() to ra cc gi tr cch u nhau trong khong  cho.
plt.plot(x, np.sin(x), color = "blue");
plt.xlabel("Bien(X)") # Label: dn nhn cho trc
plt.ylabel("Sin(X)")
plt.xlim([0,4])
plt.ylim([-0.7, 0.6])
```

```python
x = np.linspace(0, 10, 1000)  #  to ra mt mng x cha 1000 im nm trong khong t 0 n 10. np.linspace() to ra cc gi tr cch u nhau trong khong  cho.
plt.plot(x, np.sin(x), color = "blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color = "red", label="cos(x)");
plt.xlabel("Bien(X)") # Label: dn nhn cho trc
#plt.ylabel("Sin(X)")
#plt.xlim([0,4])
plt.legend();
```

### Object-Oriented API

```python
#1. Prepare data
x= [1,2,3,4]
y = [11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(5,5)) #figure size = width & height of the plot

#3. Plot the data
ax.plot(x,y);

#4. Customize the plot
ax.set(title="A simple plot",
      xlabel="x-axis",
      ylabel="y-axis");
```

### Scatter


```python
#Pylot API
plt.scatter(x, np.exp(x));   #y=e^(x)
```

```python
# OO API
#Prepare new data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

fig, ax = plt.subplots()
ax.scatter(x, y);
```

```python
# OO API
#Prepare new data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s= sizes, c = colors, cmap='viridis', alpha=0.3);  
#  s dng bn  mu 'viridis', v  trong sut (alpha) l 0.3.
fig.colorbar(img1); # Thm thanh mu sc bn cnh biu  
```

### Bar
* Vertical
* Horizontal

```python
# Prepare data from a python dictionary 
soft_drink_prices = {"Coke": 10,
                    "Pepsi": 12,
                    "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color='skyblue');
# key l tn sp and values l gi
ax.set(title="Bach Hoa Xanh", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()), color='skyblue');

# barh lt ngc biu  thnh ngang
```

### Subplots


```python
# option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10, 5))

# plot data each axis 
ax1.plot(x, x/2);
ax2.plot(x, x/2);
ax2.plot(x, x-2);

ax3.scatter(np.random.random(10), np.random.random(10));
ax4.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color='skyblue');
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#Plot data to each axis

ax[0,0].plot(x, x/2);
ax[0,1].plot(x, x/2);
ax[0,1].plot(x, x-2);

ax[1,0].scatter(np.random.random(10), np.random.random(10));
ax[1,1].bar(soft_drink_prices.keys(), soft_drink_prices.values(), color='skyblue');
```
