#### Question 3
 1. Find max and min value in 2-D array: ([[34,43,73],[82, 22, 12],[53, 94, 66]]) 
 2. Output position index of that max and min value

```python
# 1. C1
array = np.array([[34,43,73],[82, 22, 12],[53, 94, 66]])

max_value = array[0][0]
min_value = array[0][0]

for row in array:
    for value in row:
        if value > max_value:
            max_value = value
        if value < min_value:
            min_value = value
            
print("Array origin: \n", array) 
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
#1 C2
array = np.array([[34,43,73],[82, 22, 12],[53, 94, 66]])

max_value = np.max(array)
min_value = np.min(array)

print("Array origin: \n", array) 
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
# 2 Cach1
array = np.array([[34,43,73],[82, 22, 12],[53, 94, 66]])

max_value = array[0][0]
min_value = array[0][0]

max_position = (0, 0)
min_position = (0, 0)

for i, row in enumerate(array):
    for j, value in enumerate(row):
        if value > max_value:
            max_value = value
            max_position = (i, j)
        if value < min_value:
            min_value = value
            min_position = (i, j)

print("Array origin: \n", array)
print("Max value is: ", max_value)
print("index max: ", max_position)
print("Min value is: ", min_value)
print("index min: ", min_position)
       
```

```python
# cach 2

array = np.array([[34,43,73],[82, 22, 12],[53, 94, 66]])

max_index = np.unravel_index(np.argmax(array), array.shape) # np.unravel_index()  chuyn ch s dng s nguyn thnh dng (row, column) tng ng
min_index = np.unravel_index(np.argmin(array), array.shape) # numpy.argmax()  tm ch s ca gi tr ln nht , tt vi min  
print("Maximum value:", array[max_index])
print("Position of maximum value:", max_index)
print("Minimum value:", array[min_index])
print("Position of minimum value:", min_index)
```

#### Question 6
Write a function to check which element is prime in an input matrix

```python

def is_prime(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

def prime_elements(matrix):
    prime_indices = np.frompyfunc(is_prime, 1, 1)  # Hm numpy.frompyfunc() cho php bn to ra mt ufunc t mt hm Python thng thng v s dng ufunc ny  p dng hm  cho vo cc phn t ca mt hoc nhiu mng.
    prime_matrix = prime_indices(matrix).astype(bool)
    return prime_matrix

input_matrix = np.array([[2, 4, 7], [11, 16, 19], [22, 29, 31]])

prime_matrix = prime_elements(input_matrix)

print(prime_matrix)
```

#### Question 7
Multiply the following 2 matrices:

([[1, 3, 4], [-2, 6, 0], [-5, 7, 2]]), 

([[2, 3, 4], [-1, -2, -3], [0, 4, -4]])

```python
# Cach 1
arr1 = np.array([[1, 3, 4], [-2, 6, 0], [-5, 7, 2]])
arr2 = np.array([[2, 3, 4], [-1, -2, -3], [0, 4, -4]])

print("arr1 * arr2 = \n", arr1 @ arr2)
print()
print("arr2 * arr1 = \n", arr2 @ arr1)
```

```python
# Cach 2
arr1 = np.array([[1, 3, 4], [-2, 6, 0], [-5, 7, 2]])
arr2 = np.array([[2, 3, 4], [-1, -2, -3], [0, 4, -4]])

print("arr1 * arr2 = \n", arr1.dot(arr2))
print()
print("arr 2 * arr1  \n", arr2.dot(arr1))
```
