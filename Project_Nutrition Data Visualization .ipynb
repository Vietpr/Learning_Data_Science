## Project: Nutrition Data Visualization on 80 cereal products

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
cereal_df = pd.read_csv("cereal.csv")
```

```python
cereal_df.head()
```

```python
#irrelevant fields

fields = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis=1)
cereal_df_new.head()
```

```python
# corr()---> just colums (number, số không lấy string)
cereal_corr = cereal_df_new.corr(numeric_only=True); # get correlation data
cereal_corr
```

```python
# ones_like can built a matrix of booleans( true, false) with the same shape as our data
ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask = mask);
```

```python
# when removing the upper triangle, here are still two empty cells in our matrix(calories & vitamin)
adjusted_mask = mask[1:, :-1];
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask);
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
           annot = True, fmt=".2f", cmap="Blues",
           vmin=-1, vmax=1,
           linecolor='white', linewidths=0.5);
```

```python
fig, ax = plt.subplots(figsize=(10,8))

# That method uses HUSL colors, so you need hue, saturation, and lightness.
# I used hsluv.org to select the colors of this chart.
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
           annot = True, fmt=".2f", cmap=cmap,
           vmin=-1, vmax=1,
           linecolor='white', linewidths=0.5);

yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0);
ax.set_xticklabels(xticks, rotation=90);
title = 'CORRELATION MATRIX'
ax.set_title(title, loc='left', fontsize=18);
```
