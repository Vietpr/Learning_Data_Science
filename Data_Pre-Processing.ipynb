## Data Implutation (Missing Data Replacement)
### Datases often have missing values and this can cause problems for machine learning alogorithms. It is considered good practice to identify and replace missing values in each column of your dataset prior to performing predictive modelling. This method of missing data to replacement is reffered to as data imputation

```python
data_df.info()
```

```python
for col in data_df.columns:
    missing_data = data_df[col].isna().sum()
    missing_percent = missing_data/len(data_df)*100
    print(f"Column {col}: has {missing_percent}% missing data")
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.heatmap(data_df.isna(), cmap="Blues", cbar=False, yticklabels=False);
```

```python
data_df.head()
```

```python
x = data_df.iloc[:,:-1].values
```

```python
x
```

```python
y = data_df.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.impute import SimpleImputer

#Create an instance of class simpleimputer: np.nan is the empty value in the dataset
imputer = SimpleImputer(missing_values=np.nan, strategy = "mean")

imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```python
x
```

## Encode Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder="passthrough")
X = ct.fit_transform(x)
```
