# Data Manipulation

```python
df.item_price.dtype
# check kiểu dữ liệu của cột item_price
```

```python
df.item_price = df.item_price.apply(lambda x: float(x.replace("$", "")))

# aplly() :sử dụng để áp dụng một hàm trên toàn bộ dòng hoặc cột của DataFrame.
```

```python
df.head()
```

```python
df["total_price"] = df["quantity"] * df["item_price"]

# tạo thêm cột total_price
```

```python
df.head()
```

```python
# tính tổng giá tiền trong cái dataset
# gán 1 biến là revenue
# hàm sum() tính tổng tất cả giá tiền và cộng lại
revenue = df["total_price"].sum()
print(revenue)
```

# Group By


```python
# Question: What was the most-ordered item? (Mặt hàng được đặt hàng nhiều nhất là gì?)
# group by: nhóm các dòng dữ liệu theo giá trị của một hoặc nhiều cột. 
c = df.groupby("item_name") ["quantity"].sum()
c.sort_values(ascending = False).head(5)

# sort_values(): xap xep increasing()
# sort_values(ascending = False) : sap xep decreasing
```
