### Project: Data Visualization _California  Cities 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()    
```

```python
# Extract Latd(latitude) = Vĩ độ and Longd(longitude) = Kinh độ 

lat, lon = cities['latd'], cities['longd']

#Plot using pylot API
plt.scatter(lon, lat);
plt.grid(True)
plt.xlabel("Longtitude")
plt.ylabel("Latitude");                
```

```python
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

#Plot using pylot API
plt.scatter(lon, lat,
           c=np.log10(population), cmap='viridis',
           s=area, linewidths=0, alpha=0.5);
plt.axis('equal')
plt.xlabel("Longtitude")
plt.ylabel("Latitude"); 
plt.colorbar(label='log$_{10}$(population)');

plt.title("California Cities: Population and Area Distribution");
```
